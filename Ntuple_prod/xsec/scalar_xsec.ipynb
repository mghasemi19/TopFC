## Compute new physics scalar and cross section dependency

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
import pandas as pd

#ttbar-charm
xsec = [0.01368, 0.0056, 0.0027, 0.0014, 0.0008, 0.0005, 0.0003, 0.0001]
gamma = [1000, 1250, 1500, 1750, 2000, 2250, 2500, 2750]

def func(x, m, a, b, c):
    return m*(1/(a*x**2+b))+c
popt, pcov = optimize.curve_fit(func, gamma, xsec,method="lm")

print(popt)
#print(pcov)

plt.figure(figsize=(7,5))
plt.plot(gamma, xsec, 'ro', label='data')
plt.plot(np.array(gamma), func(np.array(gamma), *popt), 'g--', label='fit: m=%.1f, a=%.1f, b=%.1f, c=%.1f' % tuple(popt))
plt.xlabel('$\Lambda$ (GeV)')
plt.ylabel('$\sigma_{signal}$ (pb)')
plt.title('fit_function: m*(1/(a*x**2+b))+c')
plt.legend()
```
