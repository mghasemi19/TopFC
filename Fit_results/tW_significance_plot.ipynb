```python
# Basic packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "./tWcharm_NN_opt.txt"
lines = open(filename).read().splitlines()

mu_sig = [1, 1.25, 1.5, 1.75, 2, 2.25, 2.5, 2.75, 3]
relativeBkgUncert = [0.25, 0.5, 1]
```

```python
import itertools
def split_list(lst, val):
    return [list(group) for k, group in itertools.groupby(lst, lambda x: x==val) if not k]
split_lst = split_list(lines, '')
```

```python
# AMS 0.75
pval_25 = list(map(float, split_lst[0][:9]))
pval_50 = list(map(float, split_lst[2][:9]))
pval_100 = list(map(float, split_lst[4][:9]))
zval_25 = list(map(float, split_lst[1][:9]))
zval_50 = list(map(float, split_lst[3][:9]))
zval_100 = list(map(float, split_lst[5][:9]))

print(zval_25)

plt.figure(0)
plt.plot(mu_sig, zval_25, label = "$\sigma_{B}=$25%")
plt.plot(mu_sig, zval_50, label = "$\sigma_{B}=$50%")
plt.plot(mu_sig, zval_100, label = "$\sigma_{B}=$100%")

plt.grid(True) 
plt.legend(loc="upper left", fontsize=16)

plt.xlabel("mu_sig", fontsize=14)
plt.ylabel("Z score", fontsize=14)
plt.title("ttbar_charm signal")
plt.axis([1, 3, 0, 9])

plt.figure(1)
plt.plot(mu_sig, pval_25, label = "$\sigma_{B}=$25%")
plt.plot(mu_sig, pval_50, label = "$\sigma_{B}=$50%")
plt.plot(mu_sig, pval_100, label = "$\sigma_{B}=$100%")

plt.grid(True) 
plt.legend(loc="upper right", fontsize=16)

plt.xlabel("mu_sig", fontsize=14)
plt.ylabel("p-val", fontsize=14)
plt.title("ttbar_charm signal")
plt.axis([1, 3, 0, 0.3])
```
